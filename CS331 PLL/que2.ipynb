SVD
BREAK A MATRIX INTO U. Sigma. V^T

```python
# SVD USE : The Singular-Value Decomposition, or SVD for short, is a matrix decomposition method for reducing a matrix to its constituent parts to make certain subsequent matrix calculations simpler.
# The function takes a matrix and returns the U, Sigma, and V^T elements. The Sigma diagonal matrix is returned as a vector of singular values. The V matrix is returned in a transposed form, e.g. V.T.
```

```python
import numpy as np
from scipy.linalg import svd

def calcMat_AAt(M):
    newM = np.dot(M,M.T)
    eigenvalues,eigenvectors = np.linalg.eig(newM)
    ncols = np.argsort(eigenvalues)[::-1]
    return eigenvectors[:,ncols]

def calcMat_AtA(M):
    newM = np.dot(M.T,M)
    eigenvalues,eigenvectors = np.linalg.eig(newM)
    ncols = np.argsort(eigenvalues)[::-1]
    return eigenvectors[:,ncols].T

def calcD(M):
    if(np.size(np.dot(M,M.T)))>(np.size(np.dot(M.T,M))):
        newM = np.dot(M.T,M)
    else :
        newM = np.dot(M,M.T)

    eigenvalues, eigenvectors = np.linalg.eig(newM)
    eigenvalues = np.sqrt(eigenvalues)
    return eigenvalues[::-1]
```

```python
A = np.array([[4, 2, 0], [1, 5, 6]])

Vt = calcMat_AtA(A)
U = calcMat_AAt(A)
sigma = calcD(A)
print(Vt,"\n")
print(U,"\n")
print(sigma)
U_svd, D, VT = np.linalg.svd(A)
print(VT, "\n")
print(U_svd, "\n")
print(D)
```
